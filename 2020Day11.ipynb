```python
import numpy as np

# Load input
filename = 'inputDay11.txt'
# filename = 'inputDay11test.txt'
# curIn = np.loadtxt(filename)

# Start by loading the data
inputFile = open(filename)
inputs = []
inputs = [s for s in inputFile.read().splitlines()]

print('Done loading input')
```

```python
# Cellular Automata
# If a seat is empty (L) and there are no occupied seats adjacent to it, the seat becomes occupied.
# If a seat is occupied (#) and four or more seats adjacent to it are also occupied, the seat becomes empty.
# Otherwise, the seat's state does not change.

# wid = len(inputs[0])
# hei = len(inputs)
hei = len(inputs[0])
wid = len(inputs)

print(wid)
print(hei)

# Use numerical values instead
# 0: Floor
# 1: Empty
# 2: Occupied
iniState = np.zeros((wid,hei))
for k in range(0,wid):
    print(inputs[k]) 
    for j in range(0,hei):
        cur = inputs[k][j]
        if (cur == 'L'):
            iniState[k][j] = 1
print(iniState)
```

```python
def printState(state):
    for k in range(0,hei):
        curLine = ''
        for j in range(0,wid):
            curVal = int(state[k][j])
            if (curVal == 0):
                curLine = curLine+'.'
            elif (curVal == 1):
                curLine = curLine+'L'
            elif (curVal == 2):
                curLine = curLine+'#'
        print(curLine)

def getNeighbourCoordinates(x,y):
    xs = []
    ys = []
    if (x > 0):
        # West
        xs = np.append(xs,x-1)
        ys = np.append(ys,y)
        # NW
        if (y > 0):
            xs = np.append(xs,x-1)
            ys = np.append(ys,y-1)
        # SW
        if (y < hei-1):
            xs = np.append(xs,x-1)
            ys = np.append(ys,y+1)

    if (x < wid-1):
        # EAST
        xs = np.append(xs,x+1)
        ys = np.append(ys,y)
        # NE
        if (y > 0):
            xs = np.append(xs,x+1)
            ys = np.append(ys,y-1)
        # SE
        if (y < hei-1):
            xs = np.append(xs,x+1)
            ys = np.append(ys,y+1)
    # North
    if (y > 0):
        xs = np.append(xs,x)
        ys = np.append(ys,y-1)
    # South
    if (y < hei-1):
        xs = np.append(xs,x)
        ys = np.append(ys,y+1)

    xs = xs.astype(int)
    ys = ys.astype(int)

    return xs,ys


def checkNeigh(x,y,state):
    neiX,neiY = getNeighbourCoordinates(x,y)
    numNei = len(neiX)

    numOccu = 0
    numEmpty = 0
    for k in range(0,numNei):
        curNei = state[neiX[k]][neiY[k]]
        # if (curNei == '.'):
            # numEmpty = numEmpty + 1
        # elif (curNei == 'L'):
        # if (curNei == 'L'):
        #     numEmpty = numEmpty + 1
        # elif (curNei == '#'):
        #     numOccu = numOccu + 1
        if (curNei == 1):
            numEmpty = numEmpty + 1
        elif (curNei == 2):
            numOccu = numOccu + 1
    # print(numEmpty)
    # print(numOccu)
    return numEmpty,numOccu

def countOccu(state):
    numOccu = 0
    for x in range(0,wid):
        for y in range(0,hei):
            if (state[x][y] == 2):
                numOccu = numOccu + 1
    return numOccu

# testState = iniState.view()
# testState[0][0] = 5
# print(checkNeigh(0,0,testState))
# printState(testState)

```

```python
curState = iniState.copy()
```

```python
# curState = iniState.copy()
curCount = 0
# printState(iniState)

# maxSimu = 1
maxSimu = 2000
for numSimu in range(0,maxSimu):
    print('Time: '+str(numSimu))
    prevState = curState.copy()
    for x in range(0,wid):
        for y in range(0,hei):
            # Count neighbours
            # neiX,neiY = getNeighbourCoordinates(x,y)
            # print(y)
            # print(neiY)
            numEmpty,numOccu = checkNeigh(x,y,prevState)
            thisSeat = prevState[x][y]
            # If empty
            if (thisSeat == 1):
                # without neighbours
                if (numOccu == 0):
                    # Become occupied
                    curState[x][y] = 2
            # If occupied
            if (thisSeat == 2):
                # with four or more neighbours
                if (numOccu >= 4):
                    # Become empty
                    curState[x][y] = 1



    # # Print the state
    # printState(curState)

    prevCount = curCount
    curCount = countOccu(curState)

    if (prevCount == curCount):
        print('Part 1 solution found: '+str(curCount))
        break
```

```python
# Part 2

# def getVisCoordinates(x,y):
#     xs = []
#     ys = []
#     # West
#     curX = x
#     curY = y
#     # if (x > 0):

def checkVis(x,y,state):
    curX = x
    curY = y
    numOccu = 0
    numEmpty = 0
    # West
    while (curX > 0):
        curX = curX - 1
        toCheck = state[curX][curY]
        
        if (toCheck > 0):
            if (toCheck == 1):
                numEmpty = numEmpty + 1
            elif(toCheck==2):
                numOccu = numOccu + 1
            # Exit while loop
            break
    curX = x
    curY = y

    # East
    while (curX < wid - 1):
        curX = curX + 1
        toCheck = state[curX][curY]
        
        if (toCheck > 0):
            if (toCheck == 1):
                numEmpty = numEmpty + 1
            elif(toCheck==2):
                numOccu = numOccu + 1
            # Exit while loop
            break
    curX = x
    curY = y

    # North
    while (curY > 0 ):
        curY = curY - 1
        toCheck = state[curX][curY]
        
        if (toCheck > 0):
            if (toCheck == 1):
                numEmpty = numEmpty + 1
            elif(toCheck==2):
                numOccu = numOccu + 1
            # Exit while loop
            break
    curX = x
    curY = y

    # South
    while (curY < hei - 1 ):
        curY = curY + 1
        toCheck = state[curX][curY]
        
        if (toCheck > 0):
            if (toCheck == 1):
                numEmpty = numEmpty + 1
            elif(toCheck==2):
                numOccu = numOccu + 1
            # Exit while loop
            break
    curX = x
    curY = y
    

    # SouthEast
    while (curY < hei - 1 ) & (curX < wid - 1):
        # while (curX < wid - 1):
            curY = curY + 1
            curX = curX + 1
            toCheck = state[curX][curY]
            
            if (toCheck > 0):
                if (toCheck == 1):
                    numEmpty = numEmpty + 1
                elif(toCheck==2):
                    numOccu = numOccu + 1
                # Exit while loop
                # curX = hei
                curY = hei
    curX = x
    curY = y

    # SouthWest
    while (curY < hei - 1 ) & (curX > 0):
        # while (curX > 0):
            curX = curX - 1
            curY = curY + 1
            toCheck = state[curX][curY]
            # print((curX,curY))
            # print(toCheck)
            
            if (toCheck > 0):
                if (toCheck == 1):
                    numEmpty = numEmpty + 1
                elif(toCheck==2):
                    numOccu = numOccu + 1
                # Exit while loop
                # curX = 0
                curY = hei
    curX = x
    curY = y
    
    # NorthEast
    while (curY > 0) & (curX < wid - 1):
            curX = curX + 1
            curY = curY - 1
            toCheck = state[curX][curY]
            
            if (toCheck > 0):
                if (toCheck == 1):
                    numEmpty = numEmpty + 1
                elif(toCheck==2):
                    numOccu = numOccu + 1
                # Exit while loop
                # curX = wid
                curY = 0
    curX = x
    curY = y

    # NorthWest
    while (curY > 0) & (curX > 0):
            curX = curX - 1 
            curY = curY - 1
            toCheck = state[curX][curY]
            
            if (toCheck > 0):
                if (toCheck == 1):
                    numEmpty = numEmpty + 1
                elif(toCheck == 2):
                    numOccu = numOccu + 1
                # Exit while loop
                # curX = 0
                curY = 0
    curX = x
    curY = y
    

    return numEmpty,numOccu


# testState = np.copy(iniState)
# # testState[7][4] = 8
# # testState[9][2] = 2
# # print(checkVis(7,4,testState))
# # print(testState)
# numEmpty,numOccu = checkVis(0,0,testState)
# print(numEmpty,numOccu)
```
